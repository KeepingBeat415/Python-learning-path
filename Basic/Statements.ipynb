Control Flow syntax makes use of colons and indentation

The indentation system is crucial to Python and is what sets it part from other programming languages

Syntax of an *if* statement
```python
    if some_condition:
        # execute some code
    elif some_other_condition:
        # do some
    else:
        # do something else
```

Syntax of *for loop* statement
```python
    my_iterable = [1, 2, 3]
    for item_name in my_iterable:
        print(item_name)

    for char in 'Hello World':
        print(char) # H e l l o W o r l d

    mylist = [(1,2), (3,4)]
    # tuple unpacking
    # for a, b in mylist:
    for (a, b) in mylist:
        print(a)
        print(b)
        # 1 2 3 4

    dic = {'k1': 1, 'k2':2}
    for key,value in d.item():
        print(value)

Syntax of *while loop* statement
```python
    while some_boolean_condition:
        # do something
    else
        # do something different
```

break, continue, pass
- break: break out of the current closet enclosing loop
- continue: goes to the top of the closest enclosing loop
- pass: does nothing at all
  - normally avoid syntax error
```python
    for item in x:
        # comment << error, because need indentation
        pass # as placeholder
```

# List Comprehensions are a unique way of quickly creating a list with Python

mystring = 'Hello'
mylist = [letter for letter in mystring]

mylist = [num**2 for num in range(0, 11)]
# if statement with list comprehensions
mylist = [x for x in range(0, 11) if x%2==0 ]

# if else statement with list comprehensions
mylist = [x if x%2==0 else 'ODD' for x in range(0, 11)]

# nest list
mylist = [x*y for x in [2, 4, 6] for y in [1, 10, 1000]]
